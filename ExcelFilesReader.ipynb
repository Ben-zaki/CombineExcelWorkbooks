#### reading multiple excel workbooks

```python
#selecting .xlsx extension files
path = "files\*.xlsx"
```

```python
excelfiles = glob.glob(path)
```

```python
excelfiles
```

```python
type(excelfiles)
```

#### accessing  files

```python
df0 = pd.read_excel(excelfiles[0])
df0.head(3)
```

#### stacking files one on top of the other "concatenation"

```python
#create an empty dataframe
dfglobal = pd.DataFrame()
for file in excelfiles:
    dfglobal = pd.concat([dfglobal,pd.read_excel(file)],axis = 0)#,ignore_index = True
```

```python
dfglobal.head()
```

```python
len(dfglobal)
```

#### using append method to see

```python
dfappend = pd.DataFrame()
for file in excelfiles:
    dfappend = dfappend.append([pd.read_excel(file)])
```
